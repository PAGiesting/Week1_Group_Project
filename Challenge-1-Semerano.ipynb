# December Subway Info

```python
df_1 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191207.txt')
df_2 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191214.txt')
df_3 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191221.txt')
df_4 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191228.txt')

combined_csv = pd.concat([df_1, df_2, df_3, df_4])
combined_csv.to_csv('combined_csv.csv', index = False, encoding='utf-8-sig')
combined_csv.head()
```

```python
#Create a new column that stores the date and time as a single pandas datetime object.
combined_csv['DATE_TIME'] = pd.to_datetime(combined_csv.DATE + " " + combined_csv.TIME,
                 format='%m/%d/%Y %H:%M:%S')
```
